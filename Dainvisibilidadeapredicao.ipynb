## Justificativa

O problema foi tratado como um problema de **classificação supervisionada** porque o objetivo do modelo é prever uma categoria previamente conhecida a partir de características demográficas e comportamentais presentes no dataset.

Neste estudo, o modelo será treinado para identificar padrões associados ao Transtorno do Espectro Autista (TEA), classificando os registros entre as categorias:

Diferentemente de problemas de regressão, o objetivo deste MVP não é prever um valor numérico contínuo, mas sim determinar a qual categoria cada registro pertence com base nas variáveis disponíveis.

A escolha pela classificação supervisionada também se justifica pela presença de um rótulo conhecido no dataset, permitindo que os algoritmos aprendam padrões a partir de exemplos previamente classificados.

Além disso, técnicas de classificação permitem comparar diferentes modelos de Machine Learning utilizando métricas apropriadas, como:

## 2.1 Funções auxiliares

Nesta seção são definidas funções reutilizáveis para avaliação dos modelos desenvolvidos no MVP.

O objetivo é reduzir repetições de código e padronizar o cálculo das métricas utilizadas durante os experimentos.

```python
def evaluate_classification(y_true, y_pred, proba=None):
    """
    Calcula métricas básicas para classificação.
    """

    results = {
        "accuracy": accuracy_score(y_true, y_pred),
        "f1_weighted": f1_score(y_true, y_pred, average="weighted")
    }

    try:
        if proba is not None and proba.shape[1] == 2:
            results["roc_auc"] = roc_auc_score(y_true, proba[:, 1])
    except:
        results["roc_auc"] = np.nan

    return results
```

# 5. Preparação dos dados e divisão treino/teste



```python
TARGET = "class"

PROBLEM_TYPE = "classificacao"

SEED = 42

# Não existem colunas de identificação
ID_COLUMNS = []

# O dataset não possui coluna de data
DATE_COLUMN = None

# Colunas que não devem ser utilizadas na modelagem
DROP_COLUMNS = ["result"]
```

```python
columns_to_exclude = set(ID_COLUMNS + DROP_COLUMNS)

if TARGET is not None:
    columns_to_exclude.add(TARGET)

features = [c for c in df.columns if c not in columns_to_exclude]

print("Tipo de problema:", PROBLEM_TYPE)
print("Target:", TARGET)
print("Número de features:", len(features))
print("Features:", features)
```

```python
X = df[features].copy()
y = df[TARGET].copy()

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=SEED,
    stratify=y
)

print("Treino:", X_train.shape)
print("Teste :", X_test.shape)
```

## 5.1 Justificativa da divisão

Nesta etapa, foi definida a variável alvo (*target*) como **`Class/ASD`**, responsável por indicar a classificação do indivíduo quanto à presença de indícios de Transtorno do Espectro Autista (TEA).

As variáveis preditoras (*features*) foram compostas pelas respostas ao questionário AQ-10 (`A1_Score` a `A10_Score`) e pelas características demográficas presentes no conjunto de dados. A coluna **`result`** foi removida da modelagem por representar a soma das respostas do questionário, informação diretamente relacionada à definição da variável alvo. Sua utilização poderia provocar **vazamento de dados (*data leakage*)**, comprometendo a avaliação real do desempenho dos modelos.

Como este trabalho trata de um problema de **classificação supervisionada**, o conjunto de dados foi dividido em **80% para treinamento** e **20% para teste**, utilizando `random_state = 42` para garantir a reprodutibilidade dos resultados.

Além disso, foi aplicada a **estratificação das classes** (`stratify=y`) durante a divisão dos dados, preservando a proporção entre as classes da variável alvo nos conjuntos de treinamento e teste. Essa estratégia reduz possíveis vieses decorrentes de distribuições desbalanceadas e proporciona uma avaliação mais confiável dos modelos de aprendizado de máquina.


# 6. Pré-processamento e pipeline

```python
# Identificação automática simples de colunas numéricas e categóricas.
# Revise manualmente, especialmente se houver datas, IDs, texto livre ou categorias codificadas como números.

num_cols = X_train.select_dtypes(include=np.number).columns.tolist()
cat_cols = X_train.select_dtypes(exclude=np.number).columns.tolist()

# Se houver coluna de data entre as features, remova do pipeline tabular ou transforme antes.
if DATE_COLUMN in cat_cols:
    cat_cols.remove(DATE_COLUMN)
if DATE_COLUMN in num_cols:
    num_cols.remove(DATE_COLUMN)

numeric_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocess = ColumnTransformer(transformers=[
    ("num", numeric_pipe, num_cols),
    ("cat", categorical_pipe, cat_cols)
], remainder="drop")

print("Colunas numéricas:", num_cols)
print("Colunas categóricas:", cat_cols)
```

## 6.1 Decisões de pré-processamento

O pré-processamento foi definido com o objetivo de preparar os dados para o treinamento dos modelos de aprendizado de máquina, garantindo consistência, reprodutibilidade e evitando vazamento de dados.

Embora a análise exploratória tenha indicado a ausência de valores ausentes no conjunto de dados, foi mantida uma etapa de imputação no pipeline como boa prática. Para as variáveis numéricas foi utilizada a **mediana**, por ser uma medida menos sensível a valores extremos. Para as variáveis categóricas foi utilizada a **moda** (valor mais frequente), preservando a categoria predominante caso existam dados ausentes em novos conjuntos de dados.

As variáveis numéricas foram **padronizadas** utilizando o **StandardScaler**, uma vez que apresentam escalas diferentes. A padronização favorece o desempenho de algoritmos sensíveis à escala dos atributos, como Regressão Logística, K-Nearest Neighbors (KNN) e Máquinas de Vetores de Suporte (SVM).

As variáveis categóricas foram transformadas por meio da técnica **One-Hot Encoding**, convertendo categorias em representações numéricas compatíveis com os algoritmos de aprendizado de máquina. A configuração `handle_unknown="ignore"` foi adotada para permitir que categorias não observadas durante o treinamento sejam tratadas corretamente na etapa de previsão.

A coluna **`result`** foi removida do conjunto de atributos por representar a soma das respostas do questionário AQ-10, informação diretamente relacionada à definição da variável alvo (`Class/ASD`). Sua utilização poderia gerar **vazamento de dados (*data leakage*)**, produzindo estimativas de desempenho irreais.

Não foi realizada engenharia de atributos nesta etapa, uma vez que o conjunto de dados já disponibiliza variáveis relevantes para o problema de classificação. Todas as variáveis utilizadas no treinamento estariam disponíveis no momento de uma previsão real, permitindo que o modelo seja aplicado em novos dados de forma consistente.


# 7. Baseline e modelos candidatos

```python
# === Definição de baseline e modelos candidatos ===

if PROBLEM_TYPE == "classificacao":
    baseline = Pipeline(steps=[
        ("preprocess", preprocess),
        ("model", DummyClassifier(strategy="most_frequent", random_state=SEED))
    ])

    candidates = {
        "LogisticRegression": Pipeline(steps=[
            ("preprocess", preprocess),
            ("model", LogisticRegression(max_iter=500, random_state=SEED))
        ]),
        "RandomForest": Pipeline(steps=[
            ("preprocess", preprocess),
            ("model", RandomForestClassifier(random_state=SEED))
        ])
    }

elif PROBLEM_TYPE == "regressao":
    baseline = Pipeline(steps=[
        ("preprocess", preprocess),
        ("model", DummyRegressor(strategy="median"))
    ])

    candidates = {
        "Ridge": Pipeline(steps=[
            ("preprocess", preprocess),
            ("model", Ridge())
        ]),
        "RandomForestRegressor": Pipeline(steps=[
            ("preprocess", preprocess),
            ("model", RandomForestRegressor(random_state=SEED))
        ])
    }

elif PROBLEM_TYPE == "clusterizacao":
    baseline = None
    candidates = {
        "KMeans_k3": Pipeline(steps=[
            ("preprocess", preprocess),
            ("model", KMeans(n_clusters=3, random_state=SEED, n_init=10))
        ]),
        "KMeans_k5": Pipeline(steps=[
            ("preprocess", preprocess),
            ("model", KMeans(n_clusters=5, random_state=SEED, n_init=10))
        ])
    }

elif PROBLEM_TYPE == "serie_temporal":
    # Para séries temporais, implemente uma baseline temporal, como "último valor observado".
    baseline = None
    candidates = {}

else:
    raise ValueError("PROBLEM_TYPE inválido.")

print("Modelos candidatos:", list(candidates.keys()))
```

## 7.1 Justificativa dos modelos

Os modelos utilizados neste trabalho foram selecionados considerando as características do problema e do conjunto de dados. Como o objetivo é prever a presença de indícios de Transtorno do Espectro Autista (TEA), trata-se de um problema de **classificação supervisionada binária**, no qual diferentes algoritmos podem ser comparados quanto ao seu desempenho.

Como referência inicial, foi utilizado o **DummyClassifier**, configurado com a estratégia `most_frequent`. Esse modelo constitui uma **baseline coerente**, pois realiza previsões sempre utilizando a classe mais frequente do conjunto de treinamento, sem aprender qualquer padrão dos dados. Dessa forma, torna-se possível verificar se os modelos treinados realmente apresentam capacidade de aprendizado superior a uma estratégia trivial.

Como modelos candidatos foram escolhidos a **Regressão Logística** e o **Random Forest**. A Regressão Logística é um algoritmo amplamente utilizado em problemas de classificação binária, oferecendo boa capacidade de generalização e facilidade de interpretação dos resultados. Além disso, seu desempenho costuma ser favorecido quando as variáveis numéricas são padronizadas, motivo pelo qual essa etapa foi incluída no pipeline de pré-processamento.

O Random Forest foi selecionado por sua capacidade de identificar relações não lineares entre as variáveis e por apresentar bom desempenho em conjuntos de dados tabulares compostos por atributos numéricos e categóricos. Esse algoritmo também é menos sensível à escala das variáveis, mas foi mantido no mesmo pipeline para garantir que todos os modelos utilizem exatamente o mesmo processo de preparação dos dados.

As variáveis categóricas foram previamente transformadas por meio de **One-Hot Encoding**, permitindo que ambos os algoritmos trabalhem adequadamente com informações categóricas. O uso de um pipeline único também garante que todas as transformações sejam aprendidas apenas com os dados de treinamento e posteriormente aplicadas ao conjunto de teste, evitando vazamento de dados (*data leakage*).

Por fim, os modelos escolhidos apresentam baixo custo computacional para o tamanho do conjunto de dados utilizado e oferecem um equilíbrio entre desempenho, interpretabilidade e facilidade de comparação, sendo adequados para a construção do MVP.

# 9. Validação e otimização de hiperparâmetros

Após a avaliação inicial dos modelos, foi realizada uma etapa de otimização de hiperparâmetros com o objetivo de verificar se seria possível obter um desempenho superior ao modelo configurado com os parâmetros padrão.

Para esta etapa, foi selecionado o algoritmo **Random Forest**, por apresentar bom desempenho em problemas de classificação e permitir diversos ajustes capazes de influenciar sua capacidade de generalização.

A otimização foi realizada por meio do **RandomizedSearchCV**, utilizando validação cruzada estratificada (**StratifiedKFold**) com cinco divisões. Essa estratégia permite avaliar diferentes combinações de hiperparâmetros em múltiplas partições do conjunto de treinamento, reduzindo a influência de uma única divisão dos dados.

Como métrica de avaliação foi utilizado o **F1-score ponderado (F1 Weighted)**, por considerar simultaneamente precisão e revocação, sendo uma medida adequada para problemas de classificação supervisionada.

Os principais hiperparâmetros avaliados foram:

- número de árvores da floresta (`n_estimators`);
- profundidade máxima das árvores (`max_depth`);
- número mínimo de amostras para divisão de um nó (`min_samples_split`).

Ao final da busca, foi selecionada a configuração que apresentou o melhor desempenho médio durante a validação cruzada.

```python
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from scipy.stats import randint

# Define the parameter distribution for RandomizedSearchCV
param_dist = {
    'model__n_estimators': randint(50, 200),
    'model__max_depth': randint(5, 20),
    'model__min_samples_split': randint(2, 10),
    'model__min_samples_leaf': randint(1, 5)
}

# Get the RandomForest pipeline from candidates (assuming 'candidates' is defined)
rf_pipeline = candidates["RandomForest"]

# Initialize StratifiedKFold for cross-validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)

# Initialize RandomizedSearchCV
search = RandomizedSearchCV(
    estimator=rf_pipeline,
    param_distributions=param_dist,
    n_iter=10,  # Number of parameter settings that are sampled
    cv=skf,
    scoring='f1_weighted',
    random_state=SEED,
    n_jobs=-1, # Use all available cores
    verbose=1
)

# Fit RandomizedSearchCV
print("Starting RandomizedSearchCV for RandomForest...")
search.fit(X_train, y_train)
print("RandomizedSearchCV finished.")

print("Melhor score na validação:", search.best_score_)
```

A busca foi realizada utilizando o algoritmo **RandomizedSearchCV**, que avalia diferentes combinações de hiperparâmetros de forma aleatória dentro de intervalos previamente definidos. Como estratégia de validação foi empregada a **validação cruzada estratificada (StratifiedKFold)** com cinco divisões, preservando a proporção das classes em cada partição do conjunto de treinamento.

# 12. Boas práticas e rastreabilidade

Durante o desenvolvimento deste MVP foram adotadas práticas para garantir a reprodutibilidade, transparência e organização do processo de construção do modelo de Machine Learning.

As principais decisões relacionadas ao pré-processamento, treinamento e avaliação foram registradas para facilitar futuras reproduções e possíveis melhorias da solução.

### Registro de decisões

| Decisão | Justificativa | Impacto esperado |
|---------|---------------|------------------|
| Utilizar `random_state = 42` (SEED) | Garantir reprodutibilidade dos experimentos. | Permite obter os mesmos resultados em diferentes execuções. |
| Remover a coluna `result` | A variável representa a pontuação total do questionário AQ-10 e possui forte relação direta com a variável alvo (`class`), podendo causar vazamento de informação (data leakage). | Evitar desempenho artificialmente elevado durante o treinamento. |
| Utilizar imputação pela mediana para variáveis numéricas | A mediana é menos sensível à presença de valores extremos. | Tornar o pré-processamento mais robusto. |
| Utilizar imputação pela moda para variáveis categóricas | Preserva a categoria mais frequente sem criar novas categorias artificiais. | Permitir o tratamento consistente de valores ausentes. |
| Aplicar padronização (`StandardScaler`) nas variáveis numéricas | Colocar todas as variáveis na mesma escala para algoritmos sensíveis à magnitude dos atributos. | Melhorar o desempenho da Regressão Logística e manter um pipeline padronizado. |
| Aplicar One-Hot Encoding nas variáveis categóricas | Converter atributos categóricos em formato numérico compatível com os algoritmos de Machine Learning. | Permitir que os modelos utilizem corretamente todas as variáveis categóricas. |
| Utilizar Pipeline e ColumnTransformer | Garantir que todas as transformações sejam aplicadas igualmente aos conjuntos de treino e teste, evitando vazamento de dados. | Melhorar a organização e a reprodutibilidade do projeto. |
| Utilizar DummyClassifier como baseline | Estabelecer uma referência mínima de desempenho antes do treinamento dos modelos supervisionados. | Verificar se os modelos realmente aprenderam padrões presentes nos dados. |
| Testar Regressão Logística e Random Forest | Comparar um modelo linear e um modelo baseado em árvores de decisão. | Avaliar diferentes abordagens para o problema de classificação. |
| Otimizar o Random Forest com RandomizedSearchCV | Buscar automaticamente combinações de hiperparâmetros capazes de melhorar a capacidade de generalização do modelo. | Obter um modelo mais robusto sem elevado custo computacional. |
| Utilizar F1-score ponderado como métrica principal | Considera simultaneamente precisão e revocação, sendo mais representativo para classificação do que utilizar apenas a acurácia. | Avaliar o desempenho do modelo de forma mais equilibrada. |
